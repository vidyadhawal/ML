# Problem Statement:
    1)predict price of Mercedez benz that is 4 yr old with milege 45000
    2) predict price of BMW x5 that is 7 yr old with milege 86000
    3) tell me the score of your model

```python
import pandas as pd
```

```python
df=pd.read_csv('carprices.csv')
df
```

```python
# as we can see here Car model column having text values we have to convert them to numeric we canuse Pandas dummy variable or use One hot encoding
```

# Pandas Dummy variable

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
# we have delete one column to avoid dummy trap also dekete Car model which is text column
final=merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y=final['Sell Price($)']
y
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

1) predict price of a mercedez benz that is 4 yr old with milege 45000

```python
model.predict([[45000,4,0,0]])
```

2) predict price of BMW x5 that is 7 yr old with milege 86000

```python
model.predict([[86000,7,0,1]])
```

# Using OneHotEncoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle
```

```python
dfle['Car Model']=le.fit_transform(dfle['Car Model'])
dfle
```

```python
X=dfle[['Car Model','Mileage','Age(yrs)']].values
X
```

```python
y=dfle['Sell Price($)']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(categorical_features=[0])
ohe
```

```python
X=ohe.fit_transform(X).toarray()
X
```

```python
X=X[:,1:]
X
```

```python
model.fit(X,y)
```
